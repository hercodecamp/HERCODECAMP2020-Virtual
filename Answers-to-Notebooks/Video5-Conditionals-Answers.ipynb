## If Keyword
The syntax for an if statement is the following: 
```
if ([conditional statement]):
    do something cool
```
where we replace `[conditional statement]` with any boolean expression we want. Let's try an example.

````
x = 5

if x == 5:
  print(x)
  print("x was five")
````

Be careful! Any unindented parts will be executed even if the condition is not met! Here's an example of mistaken spacing.


````
x = 10

if x == 5:
  print(x)
print("x was five")
````

It printed out "X was 5", which isn't what we wanted. 

## Else Keyword 

Now, what if X isn't 5, but we wanted to let the user know that by printing something else? We use the `else` clause as part of the if statement. The general formula looks like this: 

_Note_: _The '( )' around conditional statements are not required! Sometimes they can be easier to read if the conditionals are long and complication._

```
if ([conditional statement]): 
    do something cool
else:
    do something different, but equally cool 
```

Here's an example:

````
x = 6

if x == 5:
  print("x was five")
else:
  print("x was not 5")
````

## Elif Keyword
Sometimes, you want to cover more than just 2 cases. You can use the `elif` keyword to do this. The general formula looks something like this now: 

```
if ([conditional statement 1]): 
    do something cool
elif ([conditional statement 2]):
    do something new, but equally cool 
else:
    do something new, but equally cool
```

Let's try an example where we have 3 cases: negative, positive, and neither. 

````
x = -2 

if x > 0:
  print("x was positive")
elif x < 0:
  print ("x was negative")
else:
  print("x was neither negative nor positive")
````

## Nested If Statements

Sometimes, you can put if statements in the body of another if statement, if you want to :) This is called a nested if statement. We usually use these if one condition needs to be met before checking for the second condition:

````
x = "raining"
y = "umbrella"

if x == "raining":
  if y == "umbrella":
    print("You have an umbrella because it is raining")
  else:
    print("You do not have an umbrella, even though it is raining.")
````
